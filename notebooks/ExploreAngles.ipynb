```python
%matplotlib inline
import matplotlib
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
#%pylab inline
import datetime

matplotlib.interactive(True)
```

```python
import sys
import os
sys.path.insert(0, os.path.abspath('..'))
```

```python
datadir = "/media/sf_VBox_Shared/London/accelerometer_40Hz/"
```

```python
filenames = os.listdir(datadir)
datasets = [pd.read_csv(os.path.join(datadir,fn)) for fn in filenames if os.path.isfile(os.path.join(datadir,fn))]
print(len(datasets))
```

```python
#for dataset in datasets:
#    print dataset.columns
```

```python
for dataset in datasets:
    dataset['size'] = np.sqrt(dataset['Gx']**2 + dataset['Gy']**2 + dataset['Gz']**2)
    dataset['x'] = dataset['Gx']/dataset['size']
    dataset['y'] = dataset['Gy']/dataset['size']
    dataset['z'] = dataset['Gz']/dataset['size']
```

```python
fig, axes = plt.subplots(len(datasets), figsize=(10, 50))
for i, dataset in enumerate(datasets):
    axes[i].boxplot([dataset['x'], dataset['y'], dataset['z']], labels=['x', 'y', 'z']);
```
